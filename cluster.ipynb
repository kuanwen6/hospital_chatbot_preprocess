```python
import difflib

stop_words=['雙','左','右']


s = difflib.SequenceMatcher(None, "左腹痛", "臉潮紅").ratio()
print(s)
```

```python
for i in range(10):
    for k_num in range(1,10):
        k = kcom['Kcom_'+str(k_num)].loc[i]
        if not pd.isnull(k):
            for stop in stop_words:
                k = k.replace(stop,"")
            
            for c_type in cluster:
                for sym in cluster[c_type]:
                    s = difflib.SequenceMatcher(None, k, sym).ratio()
                    if s > 0.8:
                        print(sym, end=' ')
                        print(k)
```
